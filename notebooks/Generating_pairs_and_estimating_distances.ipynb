```python
import itertools
import numpy as np
```

L1 Distance Function

```python
def l1_distance(x, y):
    return np.sum(np.abs(x - y))
```

## Generating Pairs for Sample Data Points

```python
# Sample data
data = np.array([[1, 2], [3, 4], [9, -1]])

# Generate all the pairs
pairs = list(itertools.combinations(data, 2))
```

## Pair-wise Distances

```python
for pair in pairs:
  print('L1 Distance between point X={} and Y={} is: {}'.format(
      pair[0], pair[1],l1_distance(pair[0], pair[1])))
```
