# Distance Metrics and k-NN

This notebook illustrates:
- L1 (Manhattan) distance
- L2 (Euclidean) distance
- Cosine distance
- k-Nearest Neighbors (k-NN)

Each concept includes a simple example and a visualization.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
```

## (a) L1 Distance (Manhattan Distance)

```python
# Two points in 2D space
x = np.array([1, 2])
y = np.array([4, 6])

# L1 distance
l1_distance = np.sum(np.abs(x - y))
print("L1 Distance:", l1_distance)

# Visualization
plt.figure()
plt.scatter(*x, color='blue', label='Point x')
plt.scatter(*y, color='red', label='Point y')
plt.plot([x[0], y[0]], [x[1], x[1]], 'k--')
plt.plot([y[0], y[0]], [x[1], y[1]], 'k--')
plt.title("L1 (Manhattan) Distance")
plt.legend()
plt.grid()
plt.show()
```

## (b) L2 Distance (Euclidean Distance)

```python
# L2 distance
l2_distance = np.sqrt(np.sum((x - y) ** 2))
print("L2 Distance:", l2_distance)

# Visualization
plt.figure()
plt.scatter(*x, color='blue', label='Point x')
plt.scatter(*y, color='red', label='Point y')
plt.plot([x[0], y[0]], [x[1], y[1]], 'k-')
plt.title("L2 (Euclidean) Distance")
plt.legend()
plt.grid()
plt.show()
```

## (c) Cosine Distance

```python
# Two vectors
a = np.array([1, 2, 3])
b = np.array([2, 4, 6])

# Cosine similarity
cos_sim = np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
cos_dist = 1 - cos_sim

print("Cosine Similarity:", cos_sim)
print("Cosine Distance:", cos_dist)

# Visualization (2D projection)
plt.figure()
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Vector a')
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='red', label='Vector b')
plt.xlim(0, 5)
plt.ylim(0, 5)
plt.title("Cosine Similarity (Angle Between Vectors)")
plt.legend()
plt.grid()
plt.show()
```

## (d) k-Nearest Neighbors (k-NN)

```python
# Training data
X_train = np.array([
    [1, 2],
    [2, 3],
    [3, 3],
    [6, 5],
    [7, 7]
])

y_train = np.array([0, 0, 0, 1, 1])

# k-NN model
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)

# Test point
X_test = np.array([[4, 4]])
prediction = knn.predict(X_test)

print("Predicted class:", prediction[0])

# Visualization
plt.figure()
plt.scatter(X_train[y_train == 0][:, 0], X_train[y_train == 0][:, 1], color='blue', label='Class 0')
plt.scatter(X_train[y_train == 1][:, 0], X_train[y_train == 1][:, 1], color='red', label='Class 1')
plt.scatter(X_test[:, 0], X_test[:, 1], color='green', marker='x', s=100, label='Test Point')
plt.title("k-NN Classification (k=3)")
plt.legend()
plt.grid()
plt.show()
```
